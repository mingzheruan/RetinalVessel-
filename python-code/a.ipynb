```python
import cv2      # image translation
import math
import numpy as np
def move(img):
    height, width, channels = img.shape
    emptyImage2 = img.copy()
    x=10
    y=10
    for i in range(height):
     for j in range(width):
        if i>=x and j>=y:
            emptyImage2[i,j]=img[i-x][j-y]
        else:
            emptyImage2[i,j]=(0,0,0)


    return emptyImage2


img = cv2.imread("/Users/ruanmingzhe/Desktop/icbeb/DATAS/data4/origin.bmp")

cv2.namedWindow("Image")
SaltImage=move(img)
cv2.imwrite('/Users/ruanmingzhe/Desktop/icbeb/DATAS/data4/translation.bmp',SaltImage)
cv2.imshow("Image",img)
cv2.imshow("ss",SaltImage)
cv2.waitKey(0)
```
